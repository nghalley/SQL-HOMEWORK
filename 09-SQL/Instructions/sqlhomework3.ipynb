```python
# Dependencies and Setup
# SQL Alchemy
from sqlalchemy import create_engine

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# NumPy
import numpy as np
```

```python
# Create Engine and Pass in Postgres Connection
# Setup to Connect to Database 
engine = create_engine('postgres://postgres:Samman56@@localhost:5432/Sql_db')
conn = engine.connect()
```

```python
# Query All Records in the Salaries Table
salaries_data = pd.read_sql("SELECT * FROM salaries", conn)
salaries_data.head()
```

```python
# Query All Records in the Titles Table
titles_data = pd.read_sql("SELECT * FROM titles", conn)
titles_data.head()
```

```python
# Merge Columns / Combine Data
combined_data = pd.merge(salaries_data, titles_data, on="emp_no", how="inner")
combined_data.head()
```

```python
# Groupby Title
grouped_df = combined_data.groupby("title").mean()
grouped_df
```
